```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

I first read the data obtained with the Keck telescope into the frame data

```python
data = pd.read_table('Keckdata.txt', skiprows=[0, 2], sep='\s+')
```

I then select two different sets of observations taken during the first and second intense series of observations

```python
vals = (data['JD-2.44e6'] < 16856.0) & (data['JD-2.44e6']> 16845.0)
newdata1 = data[vals]

vals = (data['JD-2.44e6'] < 16905.0) & (data['JD-2.44e6']> 16885.0)
newdata2 = data[vals]
```

And I select the three important variables for both cases

```python
bjd1 = np.array(newdata1['JD-2.44e6'], 'd')
rvs1 = np.array(newdata1['mnvel'], 'd')
error1 = np.array(newdata1['errvel'], 'd')

bjd2 = np.array(newdata2['JD-2.44e6'], 'd')
rvs2 = np.array(newdata2['mnvel'], 'd')
error2 = np.array(newdata2['errvel'], 'd')
```

Now I use the orbital ephemeris obtained from the transits to define the orbital phases, and I also define a new zero point for the BJD timestamps

```python
period = 0.6892099
tc = 2454953.1910
tc = tc-2440000.0

phasesaux = (bjd1-tc)/period
phases1, orbit = np.modf(phasesaux)
bjdlv = np.mean(bjd1)
bjdrest, bjdint = np.modf(bjdlv)
phi1 = (bjdint-tc)/period
aux, phi1 = phi1-np.modf(phi1)
shortbjd1 = bjd1-bjdint

phasesaux = (bjd2-tc)/period
phases2, orbit = np.modf(phasesaux)
bjdlv = np.mean(bjd2)
bjdrest, bjdint = np.modf(bjdlv)
phi2 = (bjdint-tc)/period
aux, phi2 = phi2-np.modf(phi2)
shortbjd2 = bjd2-bjdint


data = dict([('p1', phases1), ('t1', shortbjd1), ('p2', phases2), ('t2', shortbjd2)])
allrvs = np.append(rvs1, rvs2)
allerror = np.append(error1, error2)
```

Now I define a function to fit a simple second order polynomial with time to each segment, defined in the dictionary, with the planet signal

```python
def parabolaplanet(x, K, a1, b1, c1, a2, b2, c2):
    "Model to fit a sinusoid to the RV data"
    newphase1 = x['p1']
    newbjd1 = x['t1']
    y1 = -K*np.sin(2.0*np.pi*newphase1) + a1 + b1*newbjd1 + c1*newbjd1**2
    newphase2 = x['p2']
    newbjd2 = x['t2']
    y2 = -K*np.sin(2.0*np.pi*newphase2) + a2 + b2*newbjd2 + c2*newbjd2**2
    y = np.append(y1, y2)
    return y
```

And I obtain the best fit parameters, their uncertainties, and I plot the observations and the model

```python
params, conv = curve_fit(parabolaplanet, data, allrvs, sigma=allerror)
paramserror = np.sqrt(np.diag(conv))
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(1,2,1)
ax.scatter(shortbjd1, rvs1)
ax.errorbar(shortbjd1, rvs1, yerr=error1, fmt='o')
plt.xlim([-5, 5]) 
ax.set_xlabel('Time [BJD]')
ax.set_ylabel('Radial velocity [m/s]')

K, a1, b1, c1, a2, b2, c2 = params
Kerr, a1err, b1err, c1err, a2err, b2err, c2err  = paramserror


newbjd1 = np.linspace(-5, 5, 500)
newx1 = (newbjd1/period) + phi1

newbjd2 = np.linspace(-5, 5, 500)
newx2 = (newbjd2/period) + phi2

data = dict([('p1', newx1), ('t1', newbjd1), ('p2', newx2), ('t2', newbjd2)])

yfit = parabolaplanet(data, K, a1, b1, c1, a2, b2, c2)

num = len(newbjd1)

plt.plot(newbjd1, yfit[:num] )

ax = fig.add_subplot(1,2,2)
ax.scatter(shortbjd2, rvs2)
ax.errorbar(shortbjd2, rvs2, yerr=error2, fmt='o')
plt.xlim([-5, 5]) 
ax.set_xlabel('Time [BJD]')
ax.set_ylabel('Radial velocity [m/s]')

plt.plot(newbjd2, yfit[num:] )


```
